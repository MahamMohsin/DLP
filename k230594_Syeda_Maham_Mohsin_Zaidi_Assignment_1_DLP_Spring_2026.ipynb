## Problem Description

We implement a small feedforward neural network with:
- 2 input neurons
- 1 hidden layer with 2 neurons (sigmoid activation)
- 1 output layer with 2 neurons (sigmoid activation)

We will:
1. Perform forward propagation
2. Compute Mean Squared Error
3. Apply backpropagation
4. Update weights using gradient descent


```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)
```

```python
# Inputs
x1 = 0.05
x2 = 0.94

# Targets
t1 = 0.01
t2 = 0.99

# Weights (Input → Hidden)
w1, w2 = 0.15, 0.20
w3, w4 = 0.25, 0.30

# Weights (Hidden → Output)
w5, w6 = 0.40, 0.45
w7, w8 = 0.50, 0.55

# Biases
b1 = 0.35
b2 = 0.60

# Learning rate
eta = 0.5
```

## Forward Propagation

Compute activations of the hidden layer and output layer.


```python
# Hidden layer
z_h1 = x1*w1 + x2*w2 + b1
z_h2 = x1*w3 + x2*w4 + b1

a_h1 = sigmoid(z_h1)
a_h2 = sigmoid(z_h2)

# Output layer
z_o1 = a_h1*w5 + a_h2*w6 + b2
z_o2 = a_h1*w7 + a_h2*w8 + b2

y1 = sigmoid(z_o1)
y2 = sigmoid(z_o2)

print("Hidden activations:", a_h1, a_h2)
print("Output predictions:", y1, y2)
```

## Loss Computation

We use Mean Squared Error (MSE).


```python
E1 = 0.5 * (t1 - y1)**2
E2 = 0.5 * (t2 - y2)**2
E_total = E1 + E2

print("Total Error:", E_total)
```

## Backpropagation

Compute gradients using the chain rule.


```python
# Output layer gradients
dE_dy1 = y1 - t1
dE_dy2 = y2 - t2

dy1_dz1 = sigmoid_derivative(z_o1)
dy2_dz2 = sigmoid_derivative(z_o2)

# Gradients for weights
dE_dw5 = dE_dy1 * dy1_dz1 * a_h1
dE_dw6 = dE_dy1 * dy1_dz1 * a_h2
dE_dw7 = dE_dy2 * dy2_dz2 * a_h1
dE_dw8 = dE_dy2 * dy2_dz2 * a_h2

print("Gradients:", dE_dw5, dE_dw6, dE_dw7, dE_dw8)
```

## Weight Update

Update weights using gradient descent.


```python
w5 -= eta * dE_dw5
w6 -= eta * dE_dw6
w7 -= eta * dE_dw7
w8 -= eta * dE_dw8

print("Updated weights:")
print(w5, w6, w7, w8)
```
